```python
plt.figure(figsize=(6, 3))
sns.heatmap(df.isnull(), cbar=False, yticklabels=False)
plt.title('Null Value Heatmap')
plt.show()
```

There are no missing values in our merged dataset.

```python
df.loc[:, df.columns != 'Unnamed: 0'].describe()
```

The mean and median (50% percentile) values for frequency suggest that data distribution might be skewed since the mean is significantly higher than the median.

For nufnu, nufnu_upper, and nufnu_lower, the mean is also higher than the median, suggesting a right-skewed distribution.

The nufnu_err has a lower standard deviation compared to nufnu, nufnu_upper, and nufnu_lower, which could indicate that the errors are relatively consistent.

The IQR (interquartile range, which is the difference between the 75% and 25% percentiles) for nufnu and related columns (nufnu_upper, nufnu_lower) suggests that most data points are clustered in a lower range with some outliers or extreme values affecting the mean.

For start_time and end_time, the IQR is very narrow, suggesting that most observations were made within a relatively short period.

The redshift data seems to be more normally distributed since the mean and median are relatively close.

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['redshift'], kde=True, bins=30)
plt.title('Distribution of Redshift')
plt.xlabel('Redshift')
plt.ylabel('Frequency')
plt.show()
```

The histogram along with KDE, indicates that the redshift data is not normally distributed. The presence of multiple peaks and the asymmetrical shape of the distribution suggest a multi-modal distribution, which is a sign that there are several distinct groups of data points. This could reflect the existence of objects at varying distances from Earth, possibly representing different clusters or types of galaxies.

```python
sns.heatmap(df.loc[:, df.columns != 'Unnamed: 0'].corr(), annot=True, cmap='Blues')
```

Redshift appears to have very little correlation with most other variables, indicating that the distance of the objects does not strongly affect the other variables.

The values of nufnu, nufnu_upper, and nufnu_lower are closely related to each other, as shown by their high correlation, which is expected as they represent related measurements.

## **Distributions of some feaures and Relationship with Redshift**

```python
fig, ax = plt.subplots(ncols=2, figsize=(15, 5))

fig.suptitle('Frequency and Redshift Distribution')

sns.histplot(x='frequency', data=df, ax=ax[0], bins=30, log_scale=True)
ax[0].set_title('Log-Scaled Distribution of "Frequency"')
ax[0].set_xlabel('Frequency')
ax[0].set_ylabel('Count')

sns.scatterplot(x='redshift', y='frequency', data=df, ax=ax[1], alpha=0.3)
ax[1].set_title('Relationship of "Frequency" and "Redshift"')
ax[1].set_xlabel('Redshift')
ax[1].set_ylabel('Frequency')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

```

The distribution of "Frequency" was highly skewed, with almost all values concentrated in a narrow range near zero, that's why I applied a logarithmic scale to the x-axis of the histogram to spread out the data points and better observe the distribution across a wider range of values.

Based on this scatter plot the frequency at which blazars emit is not directly influenced by their redshift.

```python
fig, ax = plt.subplots(ncols=2, figsize=(15, 5))

fig.suptitle('Nufnu and Redshift Distribution')

sns.histplot(x='nufnu', data=df, ax=ax[0], bins=30, log_scale=True)
ax[0].set_title('Log-Scaled Distribution of "Nufnu"')
ax[0].set_xlabel('Nufnu')
ax[0].set_ylabel('Count')

sns.scatterplot(x='redshift', y='nufnu', data=df, ax=ax[1], alpha=0.3)
ax[1].set_title('Relationship of "Nufnu" and "Redshift"')
ax[1].set_xlabel('Redshift')
ax[1].set_ylabel('Nufnu')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

As redshift increases, there is less spread in the nufnu values, with points appearing to be less scattered and more concentrated in horizontal bands. However the relationship between nufnu and redshift does not show a clear pattern based on this scatter plot.

```python
fig, ax = plt.subplots(ncols=2, figsize=(15, 5))

fig.suptitle('Flag Distribution and Relationship with Redshift')

sns.countplot(x='flag', data=df, ax=ax[0])
ax[0].set_title('Distribution of "Flag"')
ax[0].set_xlabel('Flag')
ax[0].set_ylabel('Count')
ax[0].tick_params(axis='x', rotation=90)

sns.stripplot(x='redshift', y='flag', data=df, ax=ax[1], alpha=0.3)
ax[1].set_title('Relationship of "Flag" and "Redshift"')
ax[1].set_xlabel('Redshift')
ax[1].set_ylabel('Flag')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

The Det category has a significantly higher count compared to the UL category, indicating that most measurements are actual detections rather than limits set due to computational errors or other constraints.

Both "Det" and "UL" flags are present across all redshift values, suggesting that the dataset includes a mix of precise measurements and computed upper limits for objects at various distances from Earth.

```python
top_categories = df['Catalog'].value_counts().index[:10]
df['Catalog_grouped'] = df['Catalog'].where(df['Catalog'].isin(top_categories), 'Other')

fig, ax = plt.subplots(ncols=2, figsize=(15, 5))

fig.suptitle('Catalog Distribution and Relationship with Redshift')

sns.countplot(x='Catalog_grouped', data=df, ax=ax[0], order=top_categories.tolist() + ['Other'])
ax[0].set_title('Distribution of top 10 "Catalog"')
ax[0].set_xlabel('Catalog')
ax[0].set_ylabel('Count')
ax[0].tick_params(axis='x', rotation=90)

sns.stripplot(x='redshift', y='Catalog_grouped', data=df, ax=ax[1], alpha=0.5, size=2)
ax[1].set_title('Relationship of "Catalog" and "Redshift"')
ax[1].set_xlabel('Redshift')
ax[1].set_ylabel('Catalog')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

The majority of the observations in our dataset come from "NEOWISE" catalog.

The distribution of Redshift within each Catalog category varies, with some categories showing a wide range and others(e.g. ALMA, OULS, FMonLC, etc.)  showing a more narrow or specific range of Redshift values.


## **Time series of Nufnu**

```python
outside_lower = (df['nufnu'] < df['nufnu_lower']).sum()
outside_upper = (df['nufnu'] > df['nufnu_upper']).sum()

total_outside_range = outside_lower + outside_upper
total_values = len(df['nufnu'])

percentage_outside_range = (total_outside_range / total_values) * 100

percentage_outside_range
```

Approximately 25% of nufnu_err data is outside of the lower and upper bounds.

```python
df['observation_date'] = pd.to_datetime(df['start_time'], unit='s')
df.sort_values('observation_date', inplace=True)
plt.figure(figsize=(14, 6))
sns.lineplot(x='observation_date', y='nufnu', data=df)
plt.title('Time Series of Nufnu')
plt.xlabel('Observation Date')
plt.ylabel('Nufnu')
plt.show()
```

 There are periods with significant peakes in nufnu values. The data shows considerable variability over time, with nufnu values fluctuating between low and high measurements.

## **Boxplot Analysis**

```python
# Select the numeric features for which you want to create box plots, now including 'nufnu_err'
features = ['frequency', 'nufnu', 'nufnu_upper', 'nufnu_lower', 'nufnu_err', 'redshift']

# Determine the layout of the subplots grid. Adjust the number of columns if you prefer a different layout.
n_rows = len(features) // 2 + len(features) % 2
n_cols = 2

# Create a figure and axes with the specified layout
fig, axes = plt.subplots(n_rows, n_cols, figsize=(10, 3 * n_rows))

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Create a box plot for each feature
for i, feature in enumerate(features):
    sns.boxplot(x=feature, data=df, ax=axes[i])
    axes[i].set_title(f'Boxplot of {feature}')
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('Value')

# If the number of features is odd, delete the last subplot (if it's unused)
if len(features) % 2 != 0:
    fig.delaxes(axes[-1])

plt.tight_layout()
plt.show()
```

The box plots for these features show a large number of outliers, as indicated by the points beyond the whiskers. This suggests that there is significant variation in the measurements of these features, which could be characteristic of the data or could indicate measurement errors or anomalies in the data. When considering the flags within the dataset, which distinguish between detected values ('Det') and estimated upper limits ('UL'), it becomes apparent that the outliers may be partly due to the inclusion of these upper limits.

The features related to nufnu exhibit a skewed distribution with several outliers.
